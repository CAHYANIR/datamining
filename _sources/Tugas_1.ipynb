```python
# Gender code
code_Gender_for_cherbourg = "F"
code_Gender_for_southampton = "M"

# binary value
value_of_one = 1
value_of_zero = 0

def change_code_embark_to_biner(Gender):
    return value_of_one if Gender == code_Gender_for_southampton else value_of_zero
```

```python
# Update all values of 'sex' series
df["Gender"] = df["Gender"].apply(change_code_embark_to_biner)
```

```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(df, idx, series):
    return df.loc[(idx - DECREMENT_BY_ONE), series]
```

```python
def get_dissimilarity_dataset(df, series_index = [], series = []):
    first_series = get_series(df, series_index[0], series)
    second_series = get_series(df, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
get_dissimilarity_dataset(df, [1,2], ["Gender", "Marital_Status"]).T
```

```python
def count_contingency_value(df, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = df.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(df.loc[(start_index - DECREMENT_BY_ONE):(last_index - DECREMENT_BY_ONE), df.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
# d(1,2)
d_1_2 = get_dissimilarity_dataset(df, [1,2], ["Gender", "Marital_Status"])
d_1_2
```

```python
count_contingency_value(d_1_2, 1, 2)
```

```python
# d(1,3)
d_1_3 = get_dissimilarity_dataset(df, [1,3], ["Gender", "Marital_Status"])
d_1_3
```

```python
count_contingency_value(d_1_3, 1, 3)
```
